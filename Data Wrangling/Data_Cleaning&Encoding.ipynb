```python
# first country in the country col
x = []
for i in range(len(netflix)):
    letter_list = netflix.loc[i, 'country'].split(",")
    gn = [a for a in letter_list][0]
    x.append(gn)

netflix['first_country'] = x
netflix[['country','first_country']].head(20)
```

```python
# drop actor
x = []
for i in range(len(netflix)):
    letter_list = netflix.loc[i, 'cast'].split(",")
    gn = [a for a in letter_list][0]
    x.append(gn)

netflix['first_actor'] = x
netflix['first_actor'].value_counts()
```

```python
#count the number of language 

x = []
for i in range(len(netflix)):
    if netflix.loc[i, 'language']== 'nan':
        gn = 0
        x.append(0)
    else:
        letter_list = netflix.loc[i, 'language'].split(",")
        gn = [a for a in letter_list]
        x.append(len(gn))

netflix['num_language'] = x
netflix[['language','num_language']]

#netflix['num_language'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
lable = LabelEncoder()

for x in ['kind','first_country']:
    # to have a dict of class & encode 
    lable.fit(netflix[x])
    label_name_mapping = dict(zip(lable.classes_, lable.transform(lable.classes_)))
    print(label_name_mapping)
    netflix[x]=lable.fit_transform(netflix[x])

netflix
```
